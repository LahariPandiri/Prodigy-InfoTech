```python
import pandas as pd
```

```python
data = pd.read_csv("DS1.csv")
data
```

```python
import matplotlib.pyplot as plt
```

```python
# Plotting the histogram for ages (continuous variable)
plt.figure(figsize=(10, 5))
plt.hist(data['Age'], bins=10, color='pink', edgecolor='blue', alpha=0.7)
plt.title("Distribution of Ages")
plt.xlabel("Range of Age")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Bar Chart for Gender (Categorical Variable)
plt.figure(figsize=(8, 5))
Gender_counts = data['Gender'].value_counts()
Gender_counts.plot(kind='bar', color=['blue', 'pink', 'purple'], alpha=0.7)
plt.title("Distribution of Gender")
plt.xlabel("range of Gender")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```
